```python
#Tentativa 4, coordenadas esfÃ©ricas


#arange(p0, p, intervals)
theta = np.arange(0, 2*np.pi, 0.01)
phi = np.arange(0, np.pi, 0.01)
r = np.sqrt(4)

#meshgrid?
theta, phi = np.meshgrid(theta, phi)

#vinculando as coordenadas cartesianas e esfÃ©ricas
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

#criando a figura
fig = plt.figure()
ax = Axes3D(fig)

#plotando a figura
ax.plot_surface(x,y,z, color = 'red', alpha = 0.5)
ax.plot_wireframe(x,y,z, color = 'black', alpha = 0.7)
plt.show()
```
